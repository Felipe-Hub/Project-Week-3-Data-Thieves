```python
from bs4 import BeautifulSoup
from lxml import etree
import requests
import pandas as pd
```

```python
url = 'https://www.glassdoor.com/Job/jobs.htm?suggestCount=0&suggestChosen=false&clickSource=searchBtn&typedKeyword=Data&sc.keyword=Data&locT=N&locId=195&jobType='
response = requests.get(url, headers={'user-agent': 'Mozilla/5.0'})
soup = BeautifulSoup(response.content, 'lxml')
soup.prettify()
```

```python
# get company names, job positions and links for job description

companies = [element.text for element in soup.find_all(
    'div', attrs = {'class':'jobInfoItem jobEmpolyerName'})]
position = [element.text for element in soup.find_all(
    'a', attrs = {'class': 'jobLink jobInfoItem jobTitle'})]

links = [element.get('href') for element in soup.find_all('a')]
description_links = list(filter(lambda x: x.startswith('/partner') if x else False, links))
description_links
```

```python
# access each job description url to get experience level and other info

description_url = list(map(lambda x: 'https://www.glassdoor.com' + x, description_links))
descriptions_response = list(
    map(lambda x: requests.get(x, headers={
        'user-agent': 'Mozilla/5.0'}), description_url))

d_soup = list(map(lambda x: BeautifulSoup(x.content, 'lxml'), descriptions_response))
```
